Pela análise acima "acredito" que 1 seja compareceu e 0 não compareceu.
Filtrando apenas os alunos que compareceram a todas as provas.

```python
df_filtro_presenca = df_enem.loc[(df_enem['TP_PRESENCA_CN']==1) & (df_enem['TP_PRESENCA_CH']==1) & (df_enem['TP_PRESENCA_LC']==1)
                        & (df_enem['TP_PRESENCA_MT']==1)].copy()
df_filtro_presenca
```

```python
df_filtro_presenca['NOTA_FINAL'] = df_filtro_presenca[['NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_MT','NU_NOTA_REDACAO']].mean(axis=1)
df_filtro_presenca
```

```python
df_media_alta = df_filtro_presenca.loc[(df_filtro_presenca['NOTA_FINAL']>769)]
nota_max = df_media_alta['NOTA_FINAL'].max()
nota_min = df_media_alta['NOTA_FINAL'].min()
print(f'Range das Médias dos Alunos Selecionados: de {nota_min} a {nota_max}')
```

```python
df_media_alta['NOTA_FINAL'].plot(kind='hist')
```
